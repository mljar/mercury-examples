```python
import mercury as mr
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D

import numpy as np
```

```python
app = mr.App(title="Matplotlib legend location for mercury", description="plt legend", show_code=False, continuous_update=True)
```

```python
# add checkbox
lege = mr.Checkbox(value=True, label="ON/OFF Legend")
```

```python
x1 = [-2, 1, 2, 3]
y1 = [-2, 1, 2, 3]
x2 = [-2, 1, 2, 3]
y2 = [1 ,4, 5, 6]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
ax.plot(x1, y1, label='Line 1')
ax.plot(x2, y2, label='Line 2')

# Add x and y axes at zero
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# Set x and y limits
xlim = ax.get_xlim()
ylim = ax.get_ylim()
ax.set_xlim(xlim[0], xlim[1])
ax.set_ylim(ylim[0], ylim[1])

# Remove the top and right spines of the plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

if lege.value:
    # Create the legend inside the plot
    plt.legend(loc='upper left', borderaxespad=1)
else:
    # Remove legend
    ax.legend().remove()

# Show the plot
plt.show()
```

```python
x1 = [2, 1, -2, -3]
y1 = [2, 1, -2, -3]
x2 = [2, 1, -2, -3]
y2 = [1 ,4, -5, -6]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
ax.plot(x1, y1, label='Line 1')
ax.plot(x2, y2, label='Line 2')

# Add x and y axes at zero
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# Set x and y limits
xlim = ax.get_xlim()
ylim = ax.get_ylim()
ax.set_xlim(xlim[0], xlim[1])
ax.set_ylim(ylim[0], ylim[1])

# Remove the top and right spines of the plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


if place.value == 'Right Top':
    # Create the legend outside the plot
    plt.legend(borderaxespad=3, bbox_to_anchor=(1.3, 1.1), loc='upper right')
    
elif place.value == 'Left Top':
    # Create the legend outside the plot
    plt.legend(borderaxespad=3, bbox_to_anchor=(0, 1.1), loc='upper right')
    
elif place.value == 'Right Button':
        # Create the legend outside the plot
    plt.legend(borderaxespad=3, bbox_to_anchor=(1.3, 0.2), loc='upper right')

# Show the plot
plt.show()
```
